```r
library(tidyverse)
```

```r
data <- read_csv("robos.csv")
```

```r
robos <- data %>%
    select(X2:X4)
```

```r
robos <- robos[-1,]
robos <- robos[-1,]
```

```r
names(robos) = robos[1,]
robos <- robos[-1,]
```

```r
robos
```

```r
robos <-robos %>%
    rename(fecha = "FECHA DE INICIO",
           delito = "DELITO",
           mod = "MODALIDAD") %>%
    mutate(fecha = as.Date(fecha, format = '%d/%m/%Y'))%>%
    mutate(delito = as.factor(delito))%>%
    mutate(mod = as.factor(mod))
```

```r
robos <- robos %>%
    filter(delito %in% c('Robo a pasajero a bordo de microbus c/v y s/v', 'Robo a pasajero a bordo de taxi c/v', 'Robo a pasajero a bordo de taxi s/v',
                      'Robo a pasajero al interior del metro c/v y s/v', 'Robo a transeunte en via publica c/v y s/v' ))%>%
    select(fecha:delito)%>%
    mutate(delito = fct_collapse(delito, 'microbus' = c('Robo a pasajero a bordo de microbus c/v y s/v'),
                                'taxi' = c('Robo a pasajero a bordo de taxi c/v', 'Robo a pasajero a bordo de taxi s/v'),
                                'metro' = c('Robo a pasajero al interior del metro c/v y s/v'),
                                'transeunte' = c('Robo a transeunte en via publica c/v y s/v'))) %>%
    group_by(fecha) %>%
    mutate(total = table(delito)[delito])%>%
    group_by(fecha, delito, total) %>%
    summarize()
```

```r
summary(robos)
```

```r
robos_microbus <- robos %>%
    filter(delito == 'microbus')

robos_taxi <- robos %>%
    filter(delito == 'taxi')

robos_transeunte <- robos %>%
    filter(delito == 'transeunte')

robos_metro <- robos %>%
    filter(delito == 'metro')
```

## ROBOS EN TRANSPORTES PÚBLICOS

```r
par(mfrow = c(2,2))
plot(robos_metro$fecha, robos_metro$total, main = 'Robos en el metro por día', xlab = 'Fecha', ylab = 'No. de robos', col = 'orange')
plot(robos_transeunte$fecha, robos_transeunte$total, main = 'Robos a transeuntes por día', xlab = 'Fecha', ylab = 'No. de robos', col = 'yellow')
plot(robos_taxi$fecha, robos_taxi$total, main = 'Robos en taxis por día', xlab = 'Fecha', ylab = 'No. de robos', col = 'violetred')
plot(robos_microbus$fecha, robos_microbus$total, main = 'Robos en microbuses por día', xlab = 'Fecha', ylab = 'No. de robos', col = 'springgreen4')
```

```r
par(mfrow = c(2,2))
hist(robos_metro$total,      freq = FALSE, main = 'Histograma de robos en el metro', xlab = 'No. de robos', ylab = 'Densidad', col = 'orange')
hist(robos_transeunte$total, freq = FALSE, main = 'Histograma de robos a transeunte', xlab = 'No. de robos', ylab = 'Densidad', col = 'yellow')
hist(robos_taxi$total,       freq = FALSE, main = 'Histograma de robos en taxis ', xlab = 'No. de robos', ylab = 'Densidad', col = 'violetred')
hist(robos_microbus$total,   freq = FALSE, main = 'Histograma de robos en microbuses', xlab = 'No. de robos', ylab = 'Densidad', col = 'springgreen4')
```

### A primera vista parece ser que el número de robos a transeuntes por día sigue una distribución normal, respecto a las otras es posible que su distribución sea exponencial o poisson, por lo que se estimaran estas tres distribuciones para cada una de las categorías

#### Utilizaremos los estimadores de maxima verosimilitud:


Para $exp(\lambda)$:
$$\hat{\lambda} = \frac{1}{\overline{x}}$$

para $poisson(\lambda)$:
$$\hat{\lambda} = \overline{x}$$

para $normal(\mu, \sigma)$:
$$\hat{\mu} = \overline{x}$$
$$\hat{\sigma}^2 = S^2 = \frac{1}{N}\sum_{i=1}^{N} (X_{i} -\bar{X})^2$$


```r
par(mfrow = c(1,1))
m <- mean(robos_metro$total)
s <- sd(robos_metro$total)
x = seq(0,28)


hist(robos_metro$total, freq = FALSE, main = 'Estimación Poisson', xlab = 'No. de robos', ylab = 'Densidad', col = 'orange')
points(dpois(x, lambda=m), col = 'black', pch = 15)

hist(robos_metro$total, freq = FALSE, main = 'Estimacion Exponencial', xlab = 'No. de robos', ylab = 'Densidad', col = 'orange')
points(dexp(x, rate=1/m), col = 'black', pch = 15)

hist(robos_metro$total, freq = FALSE, main = 'Estimacion Normal', xlab = 'No. de robos', ylab = 'Densidad', col = 'orange')
points(dnorm(x, mean = m, sd = s), col = 'black', pch = 15)
```

```r
par(mfrow = c(1,1))
m <- mean(robos_transeunte$total)
s <- sd(robos_transeunte$total)
x = seq(0,80)

hist(robos_transeunte$total, freq = FALSE, main = 'Estimación Poisson', xlab = 'No. de robos', ylab = 'Densidad', col = 'yellow')
points(dpois(x, lambda=m), col = 'black', pch = 15)

hist(robos_transeunte$total, freq = FALSE, main = 'Estimacion Exponencial', xlab = 'No. de robos', ylab = 'Densidad', col = 'yellow')
points(dexp(x, rate=1/m), col = 'black', pch = 15)

hist(robos_transeunte$total, freq = FALSE, main = 'Estimacion Normal', xlab = 'No. de robos', ylab = 'Densidad', col = 'yellow')
points(dnorm(x, mean = m, sd = s), col = 'black', pch = 15)
```

```r
par(mfrow = c(1,1))
m <- mean(robos_taxi$total)
s <- sd(robos_taxi$total)
x = seq(0,8)

hist(robos_taxi$total, freq = FALSE, main = 'Estimación Poisson', xlab = 'No. de robos', ylab = 'Densidad', col = 'violetred')
points(dpois(x, lambda=m), col = 'black', pch = 15)

hist(robos_taxi$total, freq = FALSE, main = 'Estimacion Exponencial', xlab = 'No. de robos', ylab = 'Densidad', col = 'violetred')
points(dexp(x, rate=1/m), col = 'black', pch = 15)

hist(robos_taxi$total, freq = FALSE, main = 'Estimacion Normal', xlab = 'No. de robos', ylab = 'Densidad', col = 'violetred')
points(dnorm(x, mean = m, sd = s), col = 'black', pch = 15)
```

```r
par(mfrow = c(1,1))
m <- mean(robos_microbus$total)
s <- sd(robos_microbus$total)
x = seq(0,8)

hist(robos_microbus$total, freq = FALSE, main = 'Estimación Poisson', xlab = 'No. de robos', ylab = 'Densidad', col = 'springgreen4')
points(dpois(x, lambda=m), col = 'black', pch = 15)

hist(robos_microbus$total, freq = FALSE, main = 'Estimacion Exponencial', xlab = 'No. de robos', ylab = 'Densidad', col = 'springgreen4')
points(dexp(x, rate=1/m), col = 'black', pch = 15)

hist(robos_microbus$total, freq = FALSE, main = 'Estimacion Normal', xlab = 'No. de robos', ylab = 'Densidad', col = 'springgreen4')
points(dnorm(x, mean = m, sd = s), col = 'black', pch = 15)
```

Con la prueba Shapiro-Wilk, confirmamos que hay evidencia para asegurar que los robos a transeuntes tienen una distrinución normal:

```r
shapiro.test(robos_transeunte$total)
```

Por ultimo, bootstrapeamos la media y la desviación estandar de nuestra muestra de robos a transeuntes.
Como era de esperarse, estas dos cantidades tienen una distribución normal.

```r
library(boot)
```

```r
stat <- function(data, indices){
    d <- data[indices]
    return(mean(d))
}

results <- boot(data=robos_transeunte$total, statistic=stat, R=1000)

plot(results)
```

```r
boot.ci(boot.out = results, type = c("norm", "basic", "perc", "bca"))
```

```r
stat <- function(data, indices){
    d <- data[indices]
    return(sd(d))
}

results <- boot(data=robos_transeunte$total, statistic=stat, R=1000)

plot(results)
boot.ci(boot.out = results, type = c("norm", "basic", "perc", "bca"))
```
